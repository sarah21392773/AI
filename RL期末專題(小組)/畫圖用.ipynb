### DQN

```python
Load_File_DQN=1000
Log_DQN= np.load(f"{folder}/DQN/model/Log_DQN-{Load_File_DQN}.npy", allow_pickle=True).item()
```

```python
Reward_Avg_DQN=np.ones(len(Log_DQN["TrainReward"]))
for i in range(len(Log_DQN["TrainReward"])//50):
  Reward_Avg_DQN[i*50:i*50+50]*=float(np.array(Log_DQN["TrainReward"][i*50:i*50+50]).mean())
plt.figure(figsize=(8,5)) #設定繪圖區大小
#繪圖區的標題，設定用中文字體twfont1，字體大小15
plt.title("TrainReward vs episodes",fontsize=15)
#設定橫軸和縱軸的標題
plt.xlabel("episodes",fontsize=15)
plt.ylabel("TrainReward",fontsize=15)
plt.plot(Log_DQN["TrainReward"],"b-",label="TrainReward_DQN")
plt.plot(Reward_Avg_DQN,"r-",label="TrainReward_avg_DQN")
plt.legend()
```

```python
plt.figure(figsize=(8,5)) #設定繪圖區大小
#繪圖區的標題，設定用中文字體twfont1，字體大小15
plt.title("TestReward vs episodes",fontsize=15)
#設定橫軸和縱軸的標題
plt.xlabel("TestReward",fontsize=15)
plt.ylabel("episodes(x100)",fontsize=15)
plt.plot(Log_DQN["TestReward"],"b-",label="TestReward_DQN")
plt.legend()
```

```python
def counted(dqn):
  return {
    "max":np.max(dqn),
    "min":np.min(dqn),
    "mean":np.mean(dqn),
    "std":np.std(dqn),
  }
```

```python
stats_DQN=counted(Log_DQN["TestReward"])
stats_DoubleDQN=counted(Log_DoubleDQN["TestReward"])
stats_DeulingDQN=counted(Log_DeulingDQN["TestReward"])
stats_NatureDQN=counted(Log_NatureDQN["TestReward"])
```

```python
print("DQN Stats:",stats_DQN)
print("DoubleDQN Stats:",stats_DoubleDQN)
print("DeulingDQN Stats:",stats_DeulingDQN)
print("NatureDQN Stats:",stats_NatureDQN)
```

```python
max_reward=[stats_DQN["max"],stats_DoubleDQN["max"],stats_DeulingDQN["max"],stats_NatureDQN["max"]]
min_reward=[stats_DQN["min"],stats_DoubleDQN["min"],stats_DeulingDQN["min"],stats_NatureDQN["min"]]
mean_reward=[stats_DQN["mean"],stats_DoubleDQN["mean"],stats_DeulingDQN["mean"],stats_NatureDQN["mean"]]
std_reward=[stats_DQN["std"],stats_DoubleDQN["std"],stats_DeulingDQN["std"],stats_NatureDQN["std"]]
```

```python
x=["max_reward","min_reward","mean_reward","std_reward"]
y=[max_reward,min_reward,mean_reward,std_reward]
kinds=["DQN","DoubleDQN","DuelingDQN","NatureDQN"]
for i in range(4):
  h=y[i]
  plt.bar(x,h,label=f"{kinds[i]}",width=0.1)
plt.legend()
plt.show()
```

```python
x=["max_reward","min_reward","mean_reward","std_reward"]
# y=[max_reward,min_reward,mean_reward,std_reward]
y = []
for i in range(4):
  y.append([max_reward[i],min_reward[i],mean_reward[i],std_reward[i]])
kinds=["DQN","DoubleDQN","DuelingDQN","NatureDQN"]
x_pos=np.arange(len(x))
bar_width=0.15
plt.figure(figsize=(8,5))
plt.ylabel("reward",fontsize=15)
for i in range(len(kinds)):
  offset=i*bar_width
  h=y[i]
  plt.bar(x_pos+offset,h,label=f"{kinds[i]}",width=bar_width)
plt.xticks(x_pos+bar_width*(len(kinds)-1)/2,x)
plt.legend()
plt.savefig("rewards.png")
plt.show()

l=[max_reward,min_reward,mean_reward,std_reward]
for i in range(4):
  print(x[i],"：",l[i])
```
