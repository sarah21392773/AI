# import neccessery

```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
folder = os.getcwd()
# print(folder)
DQN_type = ["DQN", "DoubleDQN", "DuelingDQN", "PERDQN", "RainbowDQN"]
colers = ['b', 'g', 'r', 'c', 'm']
nums = 980  # 第 980
steps = 100     # 每 100 回合取平均
```

# train rewards VS. Episode

### 每回合

```python
plt.figure(figsize=(10,6))

for i in range(len(DQN_type)):
    log = np.load(f"{folder}/model/{DQN_type[i]}_model/log_{nums}.npy", allow_pickle=True).item()
    episodes = range(len(log['TrainReward']))
    plt.plot(episodes, log['TrainReward'], color=colers[i], label=DQN_type[i])

plt.xlabel("Episode")
plt.ylabel("Train Reward")
plt.title("Train Reward VS. Episodes")
plt.grid(True)
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

### 每 100 回平均

```python
plt.figure(figsize=(10,6))

for i in range(len(DQN_type)):
    log = np.load(f"{folder}/model/{DQN_type[i]}_model/log_{nums}.npy", allow_pickle=True).item()
    rewards = log['TrainReward']
    weights = np.ones(steps) / steps
    rewards_avg = np.convolve(rewards, weights, mode='valid')
    episodes_avg = range(len(rewards_avg))
    plt.plot(episodes_avg, rewards_avg, color=colers[i], label=DQN_type[i])

plt.xlabel("Episode")
plt.ylabel("Train Reward")
plt.title("Train Reward VS. Episodes")
plt.grid(True)
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```
