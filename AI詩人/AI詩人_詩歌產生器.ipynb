#產詩

## 計算詞頻總和

```python
def wordCountSum(wordCount):
    sum = 0
    for value in wordCount.values():
        sum += value
    return sum
```

## 隨機檢索一個詞

```python
def retrieveRandomWord(wordCount):

    randIndex = randint(1, wordCountSum(wordCount))
    for word, value in wordCount.items():
        randIndex -= value
        if randIndex <= 0:
            return word
```

## 建立詞典

```python
def buildWordDict(fn):
    # fn can be either a filename or a list of filenames
    d = {}
    infile =  open(fn, "r")
    body = infile.read()
    infile.close()
    addWord(body, d)
    return d
```

## 添加詞到詞典

```python
def addWord(words, wordDict):
    import jieba
    words = words.replace("？", "")
    words = jieba.lcut(words)

    for i in range(1, len(words)):
        if words[i-1] not in wordDict:
            #Create a new dictionary for this word
            wordDict[words[i-1]] = {}
        if words[i] not in wordDict[words[i-1]]:
            wordDict[words[i-1]][words[i]] = 0
        wordDict[words[i-1]][words[i]] += 1

    return wordDict
```

## 主函式

```python
def main():
    fn = '/content/MyDrive/MyDrive/Colab Notebooks/AI詩人/newdata.txt'
    wordDict = buildWordDict(fn)

    #Generate text of length 100
    MAX_LEN = 50
    initialWord = choice(list(wordDict.keys()))
    text = ""
    currentWord = initialWord
    i = 0
    while True:
        i += 1
        text += currentWord
        if currentWord not in wordDict:
            currentWord = initialWord
        currentWord = retrieveRandomWord(wordDict[currentWord])
        if currentWord == '\n' and i >= MAX_LEN: break

    print(text)
```

```python
# The basic version of "new poem" generator.
print("===== 詩 ===== \n")
main()
```
