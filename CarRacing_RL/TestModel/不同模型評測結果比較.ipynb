# import 必要套件

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as plt_font
twfont1=plt_font.FontProperties(fname="/content/drive/MyDrive/解密AI黑盒子分享/字型/kaiu.ttf")
```

```python
dir=["探索率直線s01","探索率直線s03","探索率直線s05","探索率直線s07","探索率直線s09","探索率直線s10"]
color=["r","g","b","y","m","k"]
def DATA(dir,kind):
  Log= np.load(f"/content/drive/MyDrive/科展/{dir}/TestReward.npy", allow_pickle=True).item()
  return Log[kind]
```

```python
plt.figure(figsize=(8,5))
plt.title("平均值 vs 訓練次數",fontproperties=twfont1,fontsize=20)
plt.xlabel("訓練次數",fontproperties=twfont1,fontsize=20)
plt.ylabel("平均值",fontproperties=twfont1,fontsize=20)
for i in range(len(dir)):
  plt.plot(DATA(dir[i],"Mean"),color[i]+"-",label=dir[i])
plt.legend(prop=twfont1)
```

```python
plt.figure(figsize=(8,5))
plt.title("最大值 vs 訓練次數",fontproperties=twfont1,fontsize=20)
plt.xlabel("訓練次數",fontproperties=twfont1,fontsize=20)
plt.ylabel("最大值",fontproperties=twfont1,fontsize=20)
for i in range(len(dir)):
  plt.plot(DATA(dir[i],"Max"),color[i]+"-",label=dir[i])
plt.legend(prop=twfont1)
```

```python
plt.figure(figsize=(8,5))
plt.title("最小值 vs 訓練次數",fontproperties=twfont1,fontsize=20)
plt.xlabel("訓練次數",fontproperties=twfont1,fontsize=20)
plt.ylabel("最小值",fontproperties=twfont1,fontsize=20)
for i in range(len(dir)):
  plt.plot(DATA(dir[i],"Min"),color[i]+"-",label=dir[i])
plt.legend(prop=twfont1)
```

```python
plt.figure(figsize=(8,5))
plt.title("標準差 vs 訓練次數",fontproperties=twfont1,fontsize=20)
plt.xlabel("訓練次數",fontproperties=twfont1,fontsize=20)
plt.ylabel("標準差",fontproperties=twfont1,fontsize=20)
for i in range(len(dir)):
  plt.plot(DATA(dir[i],"Std"),color[i]+"-",label=dir[i])
plt.legend(prop=twfont1)
```
